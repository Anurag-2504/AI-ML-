```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Q1. For the following dataset generated using NumPy:
# np. random.normal (loc=70, scale=10 ,size=200)
# • Plot a histogram of scores
# • Draw a vertical line for the mean & try to interpret the results.

df = np.random.normal(loc=70, scale=10 ,size=200)
mean_score=np.mean(df)
```

```python
plt.hist(df,bins=20,edgecolor="white")

plt.xlabel("score")
plt.ylabel("Frequency")
plt.title("Histogram of Distribution of Scores")

plt.axvline(
    mean_score,
    linestyle="--",
    color="red",
    linewidth=1,
    label=f"Mean = {mean_score:.2f}"
)
plt.legend()
```

```python
# Q2. For the Seaborn penguins dataset, do the following:
# • Plot flipper_length_mm vs body_mass_g
# • Color by species


import seaborn as sns

# Load dataset
penguins = sns.load_dataset("penguins")

# Scatter plot

sns.scatterplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species"
)

# Labels and title
plt.title("Flipper Length vs Body Mass (Penguins)")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")

```

```python
# Q3. For the Seaborn tips dataset, do the following:
# • Create a box plot of total_bill for each day
# • Add hue="sex"
# • Identify which day has the highest median bill
# • Comment on outliers

#Loading data
tips = sns.load_dataset("tips")

#Box plot
sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex"
)

# Labels and title
plt.title("Total Bill Distribution by Day and Sex")
plt.xlabel("Day")
plt.ylabel("Total Bill")
```

```python
# Q4. For the Seaborn tips dataset, do the following:
# • Create a scatter plot of total_bill VS t i p
# • Color points by sex
# • Use different markers for smoker
# • Add a title and axis labels
# • Analyze: Do smokers tend to tip differently?

# Scatter plot
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker"
)

# Title and labels
plt.title("Total Bill vs Tip (Colored by Sex, Marker by Smoker)")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
```

```python
# Q5. For the Seaborn tips dataset, do the following:
# • Create a pivot table of average tip
# • Rows: day
# • Columns: time
# • Plot the pivot table as a heatmap
# • Annotate values
# • Interpret which day-time combination performs best


# Create pivot table
pivot_table = tips.pivot_table(
    values="tip",
    index="day",
    columns="time",
    aggfunc="mean"
)

# Heatmap
sns.heatmap(
    pivot_table,
    annot=True,
    fmt=".2f",
    cmap="YlOrRd"
)

# Title
plt.title("Average Tip by Day and Time..")
```
