```python
import pandas as pd
```

```python
df = pd.read_csv("IRIS.csv")
```

```python
# print first 10 rows
# print(df.head(10))
df.head(10)
```

```python
#print shape and data types
print(df.shape,df.dtypes)
```

```python
#print summary statistics (mean, std, min, max)
print(df.describe())
```

```python
# Select only those rows where:
# • petal_length > 4.5
# • species = "Iris-virginic"

filtered_Data = df.loc[ lambda x: (x["petal_length"]>4.5) & (x["species"]=="Iris-virginica")]
print(filtered_Data)
```

```python
 
# Group by species and compute:
# • Average sepal_length
# • Maximum petal_width
# • Standard deviation of sepal_width

result = df.groupby("species").agg(
    avg_sepal_length=("sepal_length", "mean"),
    max_petal_width=("petal_width", "max"),
    std_sepal_width=("sepal_width", "std")
)

print(result)
```

```python
# D. Create a new column:
# "petal_ratio" = petal_length / petal_width
# • For each species, find the average petal_ratio.

df["petal_ratio"] = df["petal_length"] / df["petal_width"]
avg_ratio = df.groupby("species")["petal_ratio"].mean()
print(avg_ratio)
```
