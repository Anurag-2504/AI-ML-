```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("store_data.json")
print(data)
print(type(data))
```

```python
#Correting the rating data
def clean_data(data):
    text_to_int = { "one": 1, "two": 2, "three": 3, "four": 4, "five": 5 }
    cleaned_data=[]
    unique_users= set()

    
    for user in data:
        
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_to_int):
            raw_rating = text_to_int[raw_rating]
        user["rating"] = raw_rating
        
        #Handling missing data
        raw_age = user.get("age")
        if raw_age == None:
            user["age"]= None

        #Deduplication
        if(user["name"].strip() in unique_users):
            continue
            
        unique_users.add(user["name"])
        cleaned_data.append(user)
        
    return cleaned_data
    
```

```python
data = clean_data(data)
```

```python
#Get meaningful insights from data
def get_insights(data):

    #avg rating
    total_rating=0
    for val in data:
        total_rating+=float(val["rating"])
    print(total_rating)
    print(f"Avg rating is : {total_rating/len(data)}")

    #percentage of poor rating customer
    poor_rating=0
    for val in data:
        if float(val["rating"])<3:
            poor_rating+=1

    print(f"% of poor rating customer : {poor_rating/len(data)*100}%")
    
```

```python
get_insights(data)
```

```python
def get_recommendations(data):
    recommendation = []

    for user in data:
        curr_recomm={}
        curr_recomm["name"]= user["name"]
        if( float(user["rating"]) >= 4):
            curr_recomm["brand"] = "Apple"
        else:
            curr_recomm["brand"] = "Samsung"
        recommendation.append(curr_recomm)
    return recommendation
            
```

```python
get_recommendations(data)
```
